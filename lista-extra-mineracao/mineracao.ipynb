# Mineração de texto - questão 1

### imports

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.utils import class_weight
```

### train and test data

```python
train = pd.read_csv('ReutersGrain-train.csv', escapechar='\\', quotechar="'")
test = pd.read_csv('ReutersGrain-test.csv', escapechar='\\', quotechar="'")
```

### pre-processing

```python
train['Text'] = train['Text'].str.replace('\n', ' ').str.lower()
test['Text'] = test['Text'].str.replace('\n', ' ').str.lower()
```

divide data

```python
X_train, y_train = train['Text'], train['class-att']
X_test, y_test = test['Text'], test['class-att']
```

Vectorize data with TF_IDF

```python
vectorizer = TfidfVectorizer(stop_words='english')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
print("Label distribution in trainning set:\n")
print(y_train.value_counts())

print("Label distribution in test set:")
print(y_test.value_counts())
```

## Random Forest Classification

train model

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

rf_classifier.fit(X_train_vec, y_train)

```

predict

```python
y_pred = rf_classifier.predict(X_test_vec)
```

```python
for i, (text, true_label, predicted_label) in enumerate(zip(X_test, y_test, y_pred)):
    print(f"Texto {i+1}:")
    print("Rótulo verdadeiro:", true_label)
    print("Rótulo previsto:", predicted_label)
    print("\n---\n")
```

```python
print("Acurácia:", accuracy_score(y_test, y_pred))
print("\nRelatório de Classificação:\n", classification_report(y_test, y_pred))
```
